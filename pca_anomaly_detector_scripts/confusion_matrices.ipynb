```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt; plt.style.use('dark_background')
import numpy as np
```

```python
kdd = np.array([[307, 8673], [19977, 17698]])
nsl = np.array([[8325, 1386], [7595, 5238]])
cic = np.array([[0 , 50522], [131228, 1950587]])
nids = np.array([[395096, 985415], [36996, 381727]])
```

```python
kdd_cm = ConfusionMatrixDisplay(confusion_matrix=kdd)
kdd_cm.plot()
plt.savefig('images/pca_anom_kdd.png')
plt.show()
```

```python
nsl_cm = ConfusionMatrixDisplay(confusion_matrix=nsl)
nsl_cm.plot()
plt.savefig('images/pca_anom_nsl.png')
plt.show()
```

```python
cic_cm = ConfusionMatrixDisplay(confusion_matrix=cic)
cic_cm.plot()
plt.savefig('images/pca_anom_cic.png')
plt.show()
```

```python
nids_cm = ConfusionMatrixDisplay(confusion_matrix=nids)
nids_cm.plot()
plt.savefig('images/pca_anom_nids.png')
plt.show()
```
